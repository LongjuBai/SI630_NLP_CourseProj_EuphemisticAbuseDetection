```python
import pandas as pd

# Load the CSV file
goldstandard = pd.read_csv('goldstandard.csv')

# Keep only the specified columns
goldstandard = goldstandard[['#CUE_COLUMN', 'FRAGMENT_COLUMN', 'INSTANCE_COLUMN', 'LABEL_COLUMN']]

# Replace values in CUE_COLUMN and LABEL_COLUMN
goldstandard['#CUE_COLUMN'] = goldstandard['#CUE_COLUMN'].replace('NO_CUE', 'RANDOM')
goldstandard['LABEL_COLUMN'] = goldstandard['LABEL_COLUMN'].replace('OTHER', 'NONE ABUSE')

# Save the modified data to a new CSV file
goldstandard.to_csv('modified_goldstandard.csv', index=False)

print('Modifications completed and saved as modified_goldstandard.csv')
display(goldstandard)
```
