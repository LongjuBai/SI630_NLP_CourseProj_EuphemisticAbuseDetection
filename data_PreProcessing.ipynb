```python
import pandas as pd
```

```python
# Load the data
file_path = 'ori_dataset/goldstandard.txt'
data = pd.read_csv(file_path, sep='\t')  # Assuming tab-separated values

# Convert to CSV
output_path = 'dataset/goldstandard.csv'
data.to_csv(output_path, index=False)

print(f'Data saved to {output_path}')

display(data)
```

```python
# Load the goldstandard data
goldstandard = pd.read_csv('dataset/goldstandard.csv')

# Base path for original fold files and path for saving updated fold files
orig_folds_path = 'ori_dataset/Folds/'
updated_folds_path = 'dataset/Folds/'
fold_files = ['instances.fold1.txt', 'instances.fold2.txt', 'instances.fold3.txt', 'instances.fold4.txt', 'instances.fold5.txt']

# Ensure the directory exists
import os
if not os.path.exists(updated_folds_path):
    os.makedirs(updated_folds_path)

# Process each fold file
for file in fold_files:
    # Load the fold data, skipping rows that start with the hash symbol
    fold_data = pd.read_csv(orig_folds_path + file, sep='\t', header=None, names=['CUE_PHRASE', 'INSTANCE'], comment='#')

    # Merge to match instances and add labels
    merged_data = fold_data.merge(goldstandard[['INSTANCE_COLUMN', 'LABEL_COLUMN']], left_on='INSTANCE', right_on='INSTANCE_COLUMN', how='left')

    # Drop the extra 'INSTANCE_COLUMN'
    merged_data.drop('INSTANCE_COLUMN', axis=1, inplace=True)

    # Save the updated fold data with labels as CSV
    updated_filename = file.replace('instances.', '').replace('.txt', '.csv')
    merged_data.to_csv(updated_folds_path + updated_filename, index=False)
    print(f'Updated data saved to {updated_folds_path}{updated_filename}')
    display(merged_data)
```
